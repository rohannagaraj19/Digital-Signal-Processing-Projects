# Image Upsampling Demo

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy import signal
from skimage.io import imread
from skimage.io import imsave

%matplotlib inline
```

```python
def upsample_image(U,img):
    n_rows,n_cols = img.shape
    new_img = np.zeros((n_rows*U,n_cols*U))
    
    for r in range(U*n_rows):
        for c in range(U*n_cols):
            if r % U == 0 and c % U == 0:
                new_img[r,c] = img[int(r/U),int(c/U)]
            else:
                new_img[r,c] = 0
    return new_img

def nn(U,x):
    return signal.convolve(x,np.ones(U))[:-U+1]

def linear(U,x):
    h = np.array([(U+i)/U if i < 0 else (U-i)/U for i in range(-U+1,U)])
    return signal.convolve(x,h,'same')

small_img = imread('test-image.jpg')
U=2
up = upsample_image(U,small_img)
plt.figure(figsize=(10,10))
plt.imshow(up,'gray')

up1 = upsample_image(U,small_img)
up2 = upsample_image(U,small_img)
n_rows,n_cols = up1.shape

#interpolate rows
for i in range(n_rows):
    up1[i,:] = linear(U,up1[i,:])
    up2[i,:] = nn(U,up2[i,:])
    
#interpolate columns
for i in range(n_cols):
    up1[:,i] = linear(U,up1[:,i])
    up2[:,i] = nn(U,up2[:,i])
    
plt.figure(figsize=(16,8))
plt.subplot(121)
plt.imshow(up1,'gray')
plt.title('Bilinear')
plt.subplot(122)
plt.imshow(up2,'gray')
plt.title('NN')
```
