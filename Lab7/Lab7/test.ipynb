```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy import signal
from skimage.io import imread
from skimage.io import imsave
```

```python
N = 8
x = [np.exp(-1j* (3*np.pi/4)*n) for n in range(0,N)]
print(len(x))
plt.figure()
plt.stem(np.linspace(0,N,N), np.abs(x))
plt.title("Magnitude IR")

plt.figure()
plt.stem(np.linspace(0,N,N), np.angle(x))
plt.title("Phase IR")
```

```python
X_d = np.fft.fft(x,N)
plt.figure()
plt.stem(np.linspace(0,N-1, N), np.abs(X_d))
plt.title("Mag Response of x")
print(X_d)

plt.figure()
plt.stem(np.linspace(0, N-1, N), np.angle(X_d))
plt.title("Phase response of x")
```
