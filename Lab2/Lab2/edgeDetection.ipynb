```python
# import necessary libraries for this lab
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

from skimage.io import imread

%matplotlib inline
```

```python
image = imread('puzzle.jpg', as_gray= True)
plt.figure(figsize=(10,6))
plt.imshow(image, 'gray')
plt.title('Original Image')
```

```python
h = np.array([1,-1]) #edge detection filter

rowsImage = np.zeros(image.shape)
for i in range(image.shape[0]):
    rowsImage[i,:] = signal.convolve(image[i,:], h, 'same')

colsImage = np.zeros(image.shape)
for j in range(image.shape[1]):
    colsImage[:,j] = signal.convolve(image[:,j], h, 'same')

rowsImage = rowsImage**2
colsImage = colsImage**2
edgeFinalBoss = rowsImage + colsImage
edgeFinalBoss = np.sqrt(edgeFinalBoss)

plt.figure(figsize=(10,6))
plt.imshow(edgeFinalBoss, 'gray')
```
