```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy import signal
from skimage.io import imread
from skimage.io import imsave
```

```python
x_d = [1/(1-0.5*np.exp(-1j * w)) for w in np.arange(-np.pi, np.pi, 0.1)]
plt.figure(figsize=(5,3))
plt.stem(np.arange(-np.pi, np.pi, 0.1), np.absolute(x_d))
```

```python
y_dA = [1 / (1 - 0.25 * np.exp(-1j * w)) for w in np.arange(-np.pi, np.pi, 0.1)]

fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # Adjusted figsize for larger and side-by-side subplots

# Magnitude plot
ax[0].stem(np.arange(-np.pi, np.pi, 0.1), np.absolute(y_dA))
ax[0].set_title('Magnitude of y_dA')
ax[0].set_xlabel('Frequency (rad/s)')
ax[0].set_ylabel('Magnitude')

# Phase plot
y_dB = [0.5 / (1 - 0.5 * np.exp(-1j * (w/2))) for w in np.arange(-np.pi, np.pi, 0.1)]
ax[1].stem(np.arange(-np.pi, np.pi, 0.1), np.absolute(x_d))
ax[1].set_title('Magnitude of y_dB')
ax[1].set_xlabel('Frequency (rad/s)')
ax[1].set_ylabel('Magnitude')

plt.tight_layout()
plt.show()
```
