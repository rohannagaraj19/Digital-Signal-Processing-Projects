```python
# import libraries first
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Audio
from scipy import signal
from scipy.io import wavfile
from skimage.io import imread
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
# Define the time index and frequency axis
n = np.linspace(-100, 100, 512) 
omega = np.linspace(-np.pi, np.pi, 512)

# Define the cosine signal
x = np.cos(1/8 * np.pi * n)

# Compute the FFT and shift the result
fft = np.fft.fft(x, 512)
fft_shifted = np.fft.fftshift(fft)

# Plot the magnitude of the FFT (shifted)
plt.figure()
plt.plot(omega, np.absolute(fft_shifted) / 512)
plt.title('Magnitude of the FFT of the Cosine Signal')
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
# Define the time index
n = np.arange(-100, 101)  # Use integers for the time index

# Define the extended frequency axis to capture multiple periods of the FFT
omega = np.linspace(-8 * np.pi, 8 * np.pi, 4096)  # Extend frequency axis even more

# Define the cosine signal
x = np.cos(1/8 * np.pi * n)

# Compute the FFT with more points to capture more detail
fft = np.fft.fft(x, 4096)
fft_shifted = np.fft.fftshift(fft)

# Plot the magnitude of the FFT (shifted)
plt.figure()
plt.plot(omega, np.absolute(fft_shifted) / len(n))
plt.title('Magnitude of the FFT of the Cosine Signal (Extended Frequency Axis)')
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python

# Define a range of discrete-time indices n
n = np.arange(-10, 11)

# Define the two sinusoidal signals
sin_5pi_4 = np.sin(5 * np.pi / 4 * n)
sin_neg3pi_4 = np.sin(-3 * np.pi / 4 * n)

# Plot the two signals
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.stem(n, sin_5pi_4, basefmt=" ")
plt.title(r'Sin $\frac{5\pi}{4} n$')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, sin_neg3pi_4, basefmt=" ")
plt.title(r'Sin $\frac{-3\pi}{4} n$')
plt.grid(True)

plt.tight_layout()
plt.show()

# Verify numerically that the two signals are the same
print("Are the two signals equal?", np.allclose(sin_5pi_4, sin_neg3pi_4))
```
